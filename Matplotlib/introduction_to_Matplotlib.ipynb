
# Matplotlib Example Workflow

We Want:
1.   Line Plot
2.   Scatter Plot
3. Bar Plot
4. Histogram
5. Subplots



```python
# import matplotlib get ready for plotting
%matplotlib inline
import matplotlib.pyplot as plt

# prepare data
x = [1, 2, 3, 4, 5, 6]
y = [11, 22, 33, 44, 55, 66]

# setup plot
fig, ax = plt.subplots(figsize=(5, 5)) #(width, height)

#plot data
ax.plot(x, y)

# customize plot
ax.set(title= "Example Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis")

# save & show plot
fig.savefig("example simple plot.png")
```

```python
# Make a plot from Dictionary

nut_butter_prices = {"Almond Butter": 10,
                     "Peanut Butter": 8,
                     "Cashew Butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title="Butter Prices Store",
       ylabel="Prices ($)")
```

Two Options for Subplots

```python
# Option 1

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10, 5))

# ax1, ax2, ax3, ax4 = ada 4 plot
# nrows2 ncols2 buat ngebentuk 2 column sama 2 rows

ax1.plot(x, x/2)
# bikin line plot

ax2.scatter(np.random.random(10), np.random.random(10))
# bikin scatter plot

ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
# ax.set(ylabel="Prices($)")
# bikin barchart

ax4.hist(np.random.randn(10000))
#bikin histogram
```

```python
# Option 2

fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10, 5))

#plot to each different index for axes
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1, 1].hist(np.random.randn(1000))
```

## Which one shold you use? (Pyplot vs matplotlib OO method)

*   When ploting something quickly, use the pyplot methd
*   when ploting something more advanced, use OO method



```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease['age'] > 50]
over_50
len(over_50)
```

```python
over_50.head()
```

```python
# pyplot methode

over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target')
```

```python
# oo method combination with pyplot

fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
             ax=ax)

ax.set_xlim([45, 100])
```

```python
over_50.target.values
```

```python
# OO methode from scratch
fig, ax = plt.subplots(figsize=(10, 6))

# plot the data
scatter = ax.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'])

# customize the plot
ax.set(title='Heart Disease and Cholestrol Levels',
       xlabel = 'Age',
       ylabel = 'Cholestrol')

# add a legend
ax.legend(*scatter.legend_elements(), title='Target')

# add horizontal line
ax.axhline(over_50['chol'].mean(),
           linestyle='--')
```

```python
over_50.head()
```

```python

# Subplots of chol, age, thalach

fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10, 10))

# AX0

# add data to ax0
scatter = ax0.scatter(x=over_50['age'],
                      y=over_50['chol'],
                      c=over_50['target'],
                      cmap='spring') # this change colour scheme (dots colour)

# customize ax0
ax0.set(title='Heart Disease and Cholestrol Levels',  
        ylabel='Cholestrol Levels')

# add legend ax0
ax0.legend(*scatter.legend_elements(), title='Target')

# add meanline
ax0.axhline(y=over_50['chol'].mean(),
            linestyle='--')

# set limit to index x
ax0.set_xlim([45, 100])

# AX1

# add data to ax1
scatter = ax1.scatter(x=over_50['age'],
                      y=over_50['thalach'],
                      c=over_50['target'],
                      cmap='spring')

# customize ax1
ax1.set(title='Heart Disease and Max Heart Rate',
        xlabel='Age',
        ylabel='Heart Rate (Max)')

# add a legend ax1
ax1.legend(*scatter.legend_elements(), title='Target')

# add meanline
ax1.axhline(y=over_50['thalach'].mean(),
            linestyle='--')

# set limit for index x
ax1.set_xlim([45, 100])

fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold')
```
