## Selection

While standard Python / NumPy expressions for selecting and setting are intuitive and come in handy for interactive work, for production code, we recommend the optimized pandas data access methods, .at, .iat, .loc and .iloc.

```python
df["A"]

# selecting series by column A
```

```python
df[0:3]

# selecting by slice rows
```

```python
df["20130102": "20130104"]

#  selecting by slice rows
```

SELECT BY LABEL

```python
df.loc[dates[0]]
```

```python
df.loc[:, ["A", "B"]]

# Selecting on a multi-axis by label:
```

```python
df.loc["20130102": "20130104", ["A", "B"]]

# Showing label slicing, both endpoints are included
```

```python
df.loc["20130102", ["A", "B"]]

# Reduction in the dimensions of the returned object
```

```python
df.loc[dates[0], "A"]

# For getting a scalar value
```

```python
df.at[dates[0], "A"]

# For getting fast access to a scalar (equivalent to the prior method)
```

SELECT BY POSITION

```python
df.iloc[3]

# selecting via the position of the passed integers
```

```python
df.iloc[3:5, 0:2]

# By integer slices, acting similar to NumPy/Python

# 3:5 slice dates, 0:2 slice column ABCD
```

```python
df.iloc[[1, 2, 4], [0, 2]]

# [1, 2, 4] for slicing dates
# [0, 2] for slicing column ABCD
```

```python
df.iloc[1:3, :]

# slicing rows explicityly
```

```python
df.iloc[:, 1:3]

# slicing column explicityly
```

```python
df.iloc[1, 1]

# get value
```

```python
df.iat[1, 1]

# getting fast access to a scalar (equivalent to the prior method)
```

BOOLEAN INDEXING

```python
df[df["A"] > 0]

# geting value in column A with value more than 0
```

```python
df[df > 0]

# get value more than 0
```

## GROUPING

By group by we are referring to a process involving one or more of the following steps:

1.   Splitting the data into groups based on some criteria
2.   Applying a function to each group independently
3.   Combining the results into a data structure

```python
df = pd.DataFrame({
    "A": ["foo", "bar", "foo", "bar", "foo", "bar", "foo", "foo"],
    "B": ["one", "one", "two", "three", "two", "two", "one", "three"],
    "C": np.random.randn(8),
    "D": np.random.randn(8)
})

df
```

```python
df.groupby("A").sum()
```

```python
df.groupby(["A", "B"]).sum()
```

## RESHAPING

STACK

```python
tuples = list(
    zip(
        *[
          ["bar", "bar", "baz", "baz", "foo", "foo", "qux", "qux"],
          ["one", "two", "one", "two", "one", "two", "one", "two"],
        ]
    )
)
```

```python
index = pd.MultiIndex.from_tuples(tuples, names=['First', 'Second'])
```

```python
df = pd.DataFrame(np.random.randn(8, 2), index=index, columns=["A", "B"])
df
```

```python
df2 = df[:4]
```

```python
df2
```

```python
stacked = df2.stack()
stacked
```

```python
stacked.unstack()
```

```python
stacked.unstack(1)
```

```python
stacked.unstack(0)
```
